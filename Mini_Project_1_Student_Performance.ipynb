```python
import pandas as pd

data = {
    "Student_ID": ["S001", "S002", "S003", "S004", "S005"],
    "Name": ["Taneshwar", "Aditi", "Rohan", "Meena", "Vikram"],
    "Math": [88, 45, 73, 92, 55],
    "Science": [75, 60, 90, 50, 65],
    "English": [60, 84, 75, 50, 45],
    "History" : [75, 80, 64, 70, 76],
    "Hindi" : [55, 45, -20, 50, 67]

}

df = pd.DataFrame(data)
df.to_csv("students.csv", index=False)
print("Dataset created and saved as students.csv")
```

```python
df
```

```python
def total(row):
  return row["Math"] + row["Science"] + row["English"] + row["History"] + row["Hindi"]

def average(row):
  return row["Total"] / 5

def pass_or_fail(row):
  if (
      row["Math"] >= 35 and row["Science"] >= 35 and row["English"] >= 35
      and row["History"] >= 35 and row["Hindi"] >= 35
  ):
      return "Pass"
  return "Fail"

def grade(row):
  avg = row["Average"]
  if avg >= 85:
    return "A"
  elif avg >= 70:
    return "B"
  elif avg >= 55:
    return "C"
  elif avg >= 40:
    return "D"
  else:
    return "F"

def highest_subject(row):
  subjects = ["Math", "Science", "English", "History", "Hindi"]
  scores = [row[s] for s in subjects]
  return subjects[scores.index(max(scores))]

def lowest_subject(row):
  subjects = ["Math", "Science", "English", "History", "Hindi"]
  scores = [row[s] for s in subjects]
  return subjects[scores.index(min(scores))]

```

```python
for subject in ["Math", "Science", "English", "History", "Hindi"]:
    for mark in df[subject]:
        if mark < 0 or mark > 100:
            print(f"Invalid mark found in {subject}: {mark}. Skipping entry...")
            continue

        if mark == 0:
            print(f"Zero mark found in {subject}. Stopping program...")
            break
```

```python
df["Total"] = df.apply(total, axis=1)
df["Average"] = df.apply(average, axis=1)
df["Result"] = df.apply(pass_or_fail, axis=1)
df["Grade"] = df.apply(grade, axis=1)
df["Highest_Subject"] = df.apply(highest_subject, axis=1)
df["Lowest_Subject"] = df.apply(lowest_subject, axis=1)

```
